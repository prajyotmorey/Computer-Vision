```python
# Assignment_1
 ## Prajyot Morey T21009

### Question_a
# 1. Rotation

import cv2
import matplotlib.pyplot as plt
import numpy as np
import math

Image=cv2.imread(r"C:\Users\prajy\Desktop\Image_1.png") #Taking image Input 

plt.imshow(Image) #To Show Image
plt.title("Original Image")
plt.axis('off')

#### A_1_Image Rotation

print("*****Image Rotation*****")
print("Enter The Angle Of Rotation In Degree:-")
angle=int(input())
anglerad=(math.pi*angle)/180

##### Rotation Matrix

Mat=([[math.cos(anglerad),math.sin(anglerad),0] ,
     [-math.sin(anglerad),math.cos(anglerad),0] ,
     [0,0,1]])
# Mat=([[ 2.82491202e-01, -3.35253425e-01,  2.48075534e+02],
#        [-1.27010373e-01,  3.07764905e-01,  2.42172977e+02],
#        [-7.58289678e-04, -1.09169824e-04,  1.00000000e+00]])

Mat

##Bring The Origin To Center
trans=([1,0,500] ,
       [0,1,354] ,
       [0,0,1])

trans

s=Image.shape
outx=np.zeros([s[0],s[1]])
outy=np.zeros([s[0],s[1]])

for i in range(0,707):
    for j in range(0,1000):
        new=(np.dot(Mat,[j,i,1]))
        outx[i,j]=new[0]/new[2]
        outy[i,j]=new[1]/new[2]

minoutx = np.min(outx)
minouty = np.min(outy)
maxoutx = np.max(outx)
maxouty = np.max(outy)

f=np.zeros(([int(maxouty+abs(minouty)+2),int(maxoutx+abs(minoutx)+4),3]))

for i in range(0,707):
    for j in range(0,1000):
        f[(int(outy[i,j]+abs(minouty)+1)),(int(outx[i,j]+abs(minoutx)+1)),0]=Image[i,j,0]
        f[(int(outy[i,j]+abs(minouty)+1)),(int(outx[i,j]+abs(minoutx)+1)),1]=Image[i,j,1]
        f[(int(outy[i,j]+abs(minouty)+1)),(int(outx[i,j]+abs(minoutx)+1)),2]=Image[i,j,2]

plt.imshow(f,cmap='gray')
plt.title("Rotated Image")
plt.axis("Off")
plt.show()
print("**********Angle of Rotation is*************:-",angle)

# A_2_Affine Trasformation

#Input Affine Matrix

affine_matrix=[]
for i in range(0,3):
    store=[]
    print("*******Enter",(i+1),"Row Element:-*******")
    for j in range(0,3):
        print("Enter Value of h",i,j)
        store.append(int(input()))
        #x=int(input())
    affine_matrix.append(store)

print("Affine Matrix:-")
for i in range(0,3):
    for j in range(0,3):
        print(affine_matrix[i][j], end = " ")
        #final.append(affine_matrix[i][j])
    print()

s=Image.shape
A_outx=np.zeros([s[0],s[1]])
A_outy=np.zeros([s[0],s[1]])

for i in range(0,707):
    for j in range(0,1000):
        new=(np.dot(affine_matrix,[j,i,1]))
        A_outx[i,j]=new[0]/new[2]
        A_outy[i,j]=new[1]/new[2]

A_minoutx = np.min(A_outx)
A_minouty = np.min(A_outy)
A_maxoutx = np.max(A_outx)
A_maxouty = np.max(A_outy)

f=np.zeros(([int(A_maxouty+abs(A_minouty)+2),int(A_maxoutx+abs(A_minoutx)+4),3]))

for i in range(0,707):
    for j in range(0,1000):
        f[(int(A_outy[i,j]+abs(A_minouty)+1)),(int(A_outx[i,j]+abs(A_minoutx)+1)),0]=Image[i,j,0]
        f[(int(A_outy[i,j]+abs(A_minouty)+1)),(int(A_outx[i,j]+abs(A_minoutx)+1)),1]=Image[i,j,1]
        f[(int(A_outy[i,j]+abs(A_minouty)+1)),(int(A_outx[i,j]+abs(A_minoutx)+1)),2]=Image[i,j,2]

plt.imshow(f,cmap='gray')
plt.title("Affine Transform:-")
plt.axis("Off")
plt.show()

# A_3_ Homography

#Input Affine Matrix

Homo_mat=[]
for i in range(0,3):
    store=[]
    print("*******Enter",(i+1),"Row Element:-*******")
    for j in range(0,3):
        print("Enter Value of h",i,j)
        x = input()
        if '.' in x:
            x = float(x)
        else:
            x = int(x)
        store.append(x)
        #x=int(input())
    Homo_mat.append(store)

print("Homography Matrix:-")
for i in range(0,3):
    for j in range(0,3):
        print(Homo_mat[i][j], end = " ")
        #final.append(affine_matrix[i][j])
    print()

s=Image.shape
H_outx=np.zeros([s[0],s[1]])
H_outy=np.zeros([s[0],s[1]])

for i in range(0,707):
    for j in range(0,1000):
        new=(np.dot(Homo_mat,[j,i,1]))
        H_outx[i,j]=new[0]/new[2]
        H_outy[i,j]=new[1]/new[2]

H_minoutx = np.min(H_outx)
H_minouty = np.min(H_outy)
H_maxoutx = np.max(H_outx)
H_maxouty = np.max(H_outy)

f=np.zeros(([int(H_maxouty+abs(H_minouty)+2),int(H_maxoutx+abs(H_minoutx)+4),3]))

for i in range(0,707):
    for j in range(0,1000):
        f[(int(H_outy[i,j]+abs(H_minouty)+1)),(int(H_outx[i,j]+abs(H_minoutx)+1)),0]=Image[i,j,0]
        f[(int(H_outy[i,j]+abs(H_minouty)+1)),(int(H_outx[i,j]+abs(H_minoutx)+1)),1]=Image[i,j,1]
        f[(int(H_outy[i,j]+abs(H_minouty)+1)),(int(H_outx[i,j]+abs(H_minoutx)+1)),2]=Image[i,j,2]

plt.imshow(f,cmap='gray')
plt.title("Homography Transform:-")
plt.axis("Off")
plt.show()

```
