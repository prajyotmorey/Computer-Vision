```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from PIL import Image
import math
```

```python
#Reading Images
a1  = np.mean(np.array(Image.open(r"C:\Users\prajy\Downloads\images(1) (2)\scene1.row3.col2.ppm")),2) #Right Image
a = np.mean(np.array(Image.open(r"C:\Users\prajy\Downloads\images(1) (2)\scene1.row3.col3.ppm")),2) #Left image
```

```python
n   = 5 #Filter size
ser = 20 #pixels to search 
b   = a.shape
out = np.zeros((a.shape[0],a.shape[1]))
```

```python
for i in (range(a.shape[0]-n)):
    for j in range(a.shape[1]-n):
        sub = []
        box1 = a[i:i+n,j:j+n] #box for left image
        for j1 in range(j,min(j+ser,b[1]-n)): #in 2nd image
            box2 = a1[i:i+n,j1:j1+n] #box for right image
            diff = (box2-box1).reshape(1,n**2) #diff of two box
            sub.append(np.sum(abs(diff))) #SAD
        if len(sub)!=0:
            temp     = np.argmin(np.array(sub)) #find min
            out[i,j] = np.argmin(np.array(sub)) 
        else:
            out[i,j] = temp
plt.figure()
plt.imshow(np.multiply(out/ser,255),cmap = 'gray') #Display Image
plt.colorbar()
```
